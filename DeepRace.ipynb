```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread('/content/sample_data/Rey21.jpeg')
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
result = DeepFace.analyze(img1, actions = ['race'])
result1 = DeepFace.analyze(img1, actions = ['gender'])
result2 = DeepFace.analyze(img1, actions = ['age'])
```
